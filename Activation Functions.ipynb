```python
#Linear or Identity

import matplotlib.pyplot as plt
import numpy as np

x= np.arange(-10,10)
y=x
plt.plot(x,y,c='blue')
plt.title('Linear')
plt.show()
```

```python
#Binary or Step Function

x= np.arange(-10,10)
y=[]
for i in range (x.shape[0]):
    if (x[i]>0):
        y.append(1)
    else:
        y.append(0)
plt.plot(x,y,c='green')
plt.title('Binary')
plt.show()
```

```python
#Bipolar

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    if (x[i]>0):
        y.append(1)
    else:
        y.append(-1)
plt.plot(x,y,c='red')
plt.title('Bipolar')
plt.show()
```

```python
#Sigmoid

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    y.append(1/(1+np.exp(-x[i])))
plt.plot(x,y,c='blue')
plt.title('Sigmoid')
plt.show()
```

```python
#Tanh

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    y.append((np.exp(x[i])-np.exp(-x[i]))/(np.exp(x[i])+np.exp(-x[i])))
plt.plot(x,y,c='green')
plt.title('Tanh')
plt.show()
```

```python
#Relu

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]): 
    y.append(np.maximum(x[i],0))
plt.plot(x,y,c='red')
plt.title('Relu')
plt.show()
```

```python
#Leaky Relu

x=np.arange(-10,10)
y=[]

for i in range (x.shape[0]):
    y.append(np.maximum(x[i],x[i]*0.1))
plt.plot(x,y,c='blue')
plt.title('Leaky Relu')
plt.show()
```

```python
#ELU

x=np.arange(-10,10)
y=[]
a=1
for i in range(x.shape[0]):
    if (x[i]>0):
        y.append(x[i])
    else:
        y.append((a*(np.exp(x[i])-1)))
        
plt.plot(x,y,c='green')
plt.title('ELU')
plt.show()
```

```python
#Softmax

#exp_x = np.exp(x)      return exp_x / np.sum(exp_x, axis=0)

x=np.arange(-10,10)
y=[]

for i in range(x.shape[0]):
    y.append((np.exp(x[i])/(np.sum(np.exp(x)))))
    
plt.plot(x,y,c='red')
plt.title('Softmax')
plt.show()
```
