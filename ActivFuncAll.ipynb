```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1.Sigmoid Activation function 

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sig_der(x):
    return sigmoid(x)*(1- sigmoid(x))
```

```python
x=np.linspace(10,-10,101)
out=sigmoid(x)

plt.xlabel("values")
plt.ylabel("sigmoid(values)")
plt.title("Activation Function:Sigmoid")
plt.plot(x,out,x,sig_der(x))
plt.title('Sigmoizd Activation Function & Derivative')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 2.Tanh Activation function 

```python
def tanh(x):
      return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def der_tanh(x):
      return 1 - tanh(x) * tanh(x)
```

```python
x_data=np.linspace(10,-10,101)
y_data = tanh(x_data)
dy_data = der_tanh(x_data)

plt.xlabel("values")
plt.ylabel("Tanh(values)")
plt.plot(x_data, y_data, x_data, dy_data)
plt.title("Activation Function:hyperbolic tangent & Derivative")
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 3.RELU Activation function

```python
def ReLU(x):
    data = [max(0,value) for value in x]
    return np.array(data, dtype=float)

def der_ReLU(x):
    data = [1 if value>0 else 0 for value in x]
    return np.array(data, dtype=float)
```

```python
x_data = np.linspace(-10,10,101)
y_data = ReLU(x_data)
dy_data = der_ReLU(x_data)

plt.plot(x_data, y_data, x_data, dy_data)
plt.xlabel("values")
plt.ylabel("ReLU(values)")
plt.title('ReLU Activation Function & Derivative')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 4.Leaky RELU Activation function 

```python
def lReLU(x):
    data = [max(0.01*value,value) for value in x]
    return np.array(data, dtype=float)

def der_lReLU(x):
    data = [1 if value>0 else 0.01 for value in x]
    return np.array(data, dtype=float)
```

```python

x_data = np.linspace(-10,10,101)
y_data = lReLU(x_data)
dy_data = der_lReLU(x_data)


plt.plot(x_data, y_data, x_data, dy_data)
plt.xlabel("values")
plt.ylabel("LeakyReLU(values)")
plt.title('LeakyReLU Activation Function & Derivative')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 5.Soft Max Activation function

```python
from derivative import dxdt

def softmax(t):
    return np.exp(t) / np.sum(np.exp(t))
def der_softmax(t):
    return dxdt(softmax(t), t)
```

```python
x_data=np.linspace(10,-10,80)
y_data = softmax(x_data)
dy_data= der_softmax(x_data)

plt.plot(x_data, y_data, x_data, dy_data)
plt.xlabel("values")
plt.ylabel("softmax(values)")
plt.title('softmax Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 6.Linear Activation function

```python
def linear(x):
    return x
def der_linear(x):
    return dxdt(linear(x), x)
```

```python
x = np.linspace(-10, 10,101)

plt.plot(x, linear(x), x, der_linear(x))
plt.xlabel("values")

plt.ylabel("Linear(values)")
plt.title('Linear Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 7.ELU Activation function

```python
def elu(x):
    return np.where(x < 0, 0.5 * (np.exp(x) - 1), x)

def der_elu(x):
    return np.where(x < 0, 0.5 * np.exp(x), 1)
```

```python
x = np.linspace(-10, 10, 101)

plt.plot(x, elu(x), label='f(x)')
plt.plot(x, der_elu(x), label="f''(x)")
plt.xlabel("Values")
plt.ylabel("ELU and its Derivative")
plt.title('ELU Activation Function and its Derivative')
plt.legend()
plt.grid()
plt.show()
```

# 8.SoftPlus Activation function

```python
def Softplus(x): 
    return np.log(1 + np.exp(-np.abs(x)))
def der_Softplus(x): 
    return (1/(1 + np.exp(-x)))
```

```python
x = np.linspace(-10,10,101)
plt.xlabel("values") 
plt.ylabel("Softplus(x)") 
plt.plot(x, Softplus(x), x, der_Softplus(x))
plt.title('SoftPlus Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 9.Swish Activation function

```python
def swish(x, B=1):
    return x/(1 + np.exp(-B*x))

def der_swish(x):
    
    return dxdt(swish(x), x)
```

```python
x = np.linspace(-10,10,101)
plt.xlabel("values") 
plt.ylabel("swish(x)") 
plt.plot(x, swish(x),x,der_swish(x))
plt.title('Swish Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```
